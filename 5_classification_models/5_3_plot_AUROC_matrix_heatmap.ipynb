## Plot hetamap

```python
heatdata = pd.read_csv("CRC_Fungi_AUROC_matrix.csv",index_col=0)

```

```python
heatdata = heatdata.iloc[0:6,]
heatdata.rename(index={'Average':'Avg'}, columns={'Average':'Avg'}, inplace = True)
heatdata
```

```python
font = {'family': 'Arial',
        'size': 16,
        'weight':'bold'
        }
sns.set(font_scale=1.0)
plt.rc('font',family='Arial')
fig = plt.figure(figsize = (6, 6),dpi=300)
```

```python
ax = sns.heatmap(heatdata,vmin=0.50, vmax=1.00, cmap = 'OrRd', linewidths= 0.3, annot=True, fmt='.2f',
                     annot_kws={'size':14,'weight':'bold'},cbar=True)
ax.tick_params(top=True, bottom=False, labeltop=True, labelbottom=False)
ax.set_aspect(1)
label_x = ax.get_xticklabels()
#plt.setp(label_x, rotation= -45, horizontalalignment='right',fontsize=14)
plt.setp(label_x, rotation= 0,fontsize=14)
label_y = ax.get_yticklabels()
plt.setp(label_y, rotation= 0,fontsize=14)

```

```python
fig = ax.get_figure()
fig.savefig('Plot/CRC_Fungi_AUROC_plot_heatmap.pdf', bbox_inches='tight')
plt.show()
```
